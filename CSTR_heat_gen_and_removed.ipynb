Initialization of Model Parameters :

```python

dt=2
f=1387917120         #f = (–H*V*Z)/(Ui*Ah)
Z=34930800           # Frequency factor
Y=5960.24157         # Y = E/R
```

```python
T = 27

Qr_pts = []
Qg_pts = []
T_pts = []

while T<109:
  Ca=((10)/(1+Z*exp(-Y/(273+T))))                #steady state concentraion
  Qr=(1500+5000)*T-(1500*25)-(5000*25)           # heat removed
  Qg=1500*Ca*f*(exp(-Y/(273+T)))                 # heat generated

  Qr_pts.append(Qr)
  Qg_pts.append(Qg)
  T_pts.append(T)


  T+=dt
```

Heat generation and heat removal curves

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(T_pts,Qr_pts, label ='Heat removed')
plt.plot(T_pts,Qg_pts, label ='Heat generated')
plt.legend()
plt.axvline(x=0, c="black",linewidth=0.5)
plt.axhline(y=0, c="black",linewidth=0.5)
plt.title("Heat generation and removal plot", y=-0.25)
plt.xlabel('Reactor Temperature °C')
plt.ylabel('Q (kcal/h)')
plt.ylim([0, 54000])
plt.xlim([27, 108])
plt.yticks(np.arange(0, 540005, 135000))
plt.xticks(np.arange(27, 109, 9))
```
