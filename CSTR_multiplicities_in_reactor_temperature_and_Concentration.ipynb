Initialization of Model Parameters :

```python
#n=41
dt=2
f=1387917120         #f = (–H*V*Z)/(Ui*Ah)
Z=34930800           # Frequency factor
Y=5960.24157         # Y = E/R
```

```python
T = 27

Ca_pts = []
Tj_pts = []
T_pts = []

while T<109:
  Ca=((10)/(1+Z*exp(-Y/(273+T))))                #steady state concentraion
  Tj=T+(10.0/3.0)*(T-25)-f*exp(-Y/(273+T))*Ca    #Jacket temperature

  Ca_pts.append(Ca)
  Tj_pts.append(Tj)
  T_pts.append(T)

  T+=dt
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

Steady state concentraion vs Jacket temperature

```python

plt.plot(Tj_pts,Ca_pts, label = "SS")
plt.axvline(x=0, c="black",linewidth=0.5)
plt.axhline(y=0, c="black",linewidth=0.5)
plt.title("Steady state concentration vs jacket temperature", y=-0.25)
plt.xlabel('$T_{J}$ °C')
plt.ylabel('$C_{A}$ (kmol/m\u00b3) ')
plt.ylim([1.5,9.5])
plt.xlim([3, 48])
plt.yticks(np.arange(1.5, 9.6, 0.5))
plt.xticks(np.arange(3, 49, 3))


plt.show()
```

Steady state Reactor Temperature vs Jacket temperature

```python
plt.plot(Tj_pts,T_pts, label ='')

plt.axvline(x=0, c="black",linewidth=0.5)
plt.axhline(y=0, c="black",linewidth=0.5)
plt.title("Steady state temperature vs jacket temperature", y=-0.25)
plt.xlabel('$T_{J}$ °C')
plt.ylabel('T °C')
plt.ylim([27, 108])
plt.xlim([3, 48])
plt.yticks(np.arange(27, 109, 9))
plt.xticks(np.arange(3, 49, 3))


plt.show()
```
